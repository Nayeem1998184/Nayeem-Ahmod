```python
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\Nayee\OneDrive\Desktop\matplotlib\env\df_batting.csv')
```

```python
data.columns
```

```python
data=data.drop(columns=['match','teamInnings','batsmanName','match_id'])
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Result'] =le.fit_transform(data['out/not_out'])
```

```python
data.head()
```

```python
data=data.drop(columns=['out/not_out'])
```

```python
data['sr'] = pd.to_numeric(data['SR'], errors='coerce')
```

```python
data.head()
```

```python
data=data.drop(columns=['SR'])
```

```python
data['sr'].fillna(data['sr'].median(),inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Splitting the data into train and test sets
X = data.drop(columns=['Result'])
y = data['Result']

# Make sure not to overwrite 'y_train' twice
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# Initialize and fit the Logistic Regression model
lor = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model = lor.fit(X_train, y_train)
```

```python
model
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
