```python
import matplotlib.pyplot as plt
plt.scatter(x=Crdata['Age'], y=Crdata['Balance'], color='Purple') 
plt.title('Total Balance According to Age') 
plt.xlabel('Age') 
plt.ylabel('Balance') 
plt.show()  
```

```python
categories=Crdata['Geography']
Values=Crdata['Balance']
plt.bar(categories,Values,color='green')
plt.xlabel('Country')
plt.ylabel('Total Balance')
plt.title('Balance by Geography')
```

```python
Labels=['France','Spain','Germany']
sizes=[3.11,3.00,1.53]
plt.pie(sizes,labels=Labels,autopct='%1.1f%%')
plt.title('Percentage of Total Balance by country')
```
