```python
#Transforming categorical data into numerical 

findata['gender']=findata['gender'].replace({'Female':1,'Male':2})
findata['Duration']=findata['Duration'].replace({'1-3 years':1,'3-5 years':2,'Less than 1 year':3,'More than 5 years':4})
findata['Invest_Monitor']=findata['Invest_Monitor'].replace({'Daily':1,'Monthly':2,'Weekly':3})
findata['Investment_Avenues']=findata['Investment_Avenues'].replace({'No':1,'Yes':2})
findata['Stock_Marktet']=findata['Stock_Marktet'].replace({'Yes':1,'No':2})
findata['Factor']=findata['Factor'].replace({'Locking Period':1,'Returns':2,'Risk':3})
findata['Objective']=findata['Objective'].replace({'Capital Appreciation':1,'Growth':2,'Income':3})
findata['Purpose']=findata['Purpose'].replace({'Returns':1,'Savings for Future':2,'Wealth Creation':3})
findata['Expect']=findata['Expect'].replace({"10%-20%":1,"20%-30%":2,"30%-40%":3})
findata['Avenue']=findata['Avenue'].replace({'Equity':1,'Fixed Deposits':2,'Mutual Fund':3,'Public Provident Fund':3})
findata['What are your savings objectives?']=findata['What are your savings objectives?'].replace({'Education':1,'Health Care':2,'Retirement Plan':3})
findata['Reason_Equity']=findata['Reason_Equity'].replace({'Capital Appreciation':1,'Dividend':2,'Liquidity':3})
findata['Reason_Bonds']=findata['Reason_Bonds'].replace({'Assured Returns':1,'Safe Investment':2,'Tax Incentives':3})
findata['Reason_FD']=findata['Reason_FD'].replace({'Fixed Returns':1,'High Interest Rates':2,'Risk Free':3})
findata['Reason_Mutual']=findata['Reason_Mutual'].replace({'Better Returns':1,'Fund Diversification':2,'Tax Benefits':3})
findata['Source']=findata['Source'].replace({'Financial Consultants':1,'Internet':2,
'Newspapers and Magazines':2,'Television':3})
```

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
x=findata.drop(columns='Expect')
y=findata['Expect']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=1)
```

```python
scaler=StandardScaler()
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler=scaler.transform(x_test)
lin =LinearRegression()
poly=PolynomialFeatures(degree=0)
x_poly_train=poly.fit_transform(x_train_scaler)
x_test_poly=poly.transform(x_test_scaler)
poly.fit(x_poly_train,y_train)
```

```python
lin.fit(x_poly_train,y_train)
```

```python
y_pred=lin.predict(x_test_poly)
y_pred
```

```python
msc=mean_absolute_error(x_poly_train,y_train)
```
