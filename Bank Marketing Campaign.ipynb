```python
#encoding 
from sklearn.preprocessing import LabelEncoder
```

```python
len=LabelEncoder()
```

```python
data['en_job']=len.fit_transform(data['job'])
data['en_marital']=len.fit_transform(data['marital'])
data['en_education']=len.fit_transform(data['education'])
data['en_default']=len.fit_transform(data['default'])
data['en_housing']=len.fit_transform(data['housing'])
data['en_loan']=len.fit_transform(data['loan'])
data['en_concact']=len.fit_transform(data['contact'])
data['en_month']=len.fit_transform(data['month'])
data['en_poutcome']=len.fit_transform(data['poutcome'])
data['en_result']=len.fit_transform(data['y'])
```

```python
data=data.drop(columns=['job','marital','education','default','housing','loan','contact','month','poutcome','y'])
```

```python
data.info()
```

```python
data.head()
```

```python
#training data
from sklearn.model_selection import train_test_split
X=data.drop(columns=['en_result'])
y=data['en_result']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dc=DecisionTreeRegressor()
```

```python
model=dc.fit(X_train,y_train)
```

```python
model
```

```python
y_pred=dc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
